```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
hr_dataset = pd.read_csv('hr_dataset.csv')
```

```python
# Data cleansing
```

```python
# Remove unnecessary columns
hr_dataset.drop(columns=['EmployeeCount', 'Over18', 'StandardHours'], inplace=True)
```

```python
# Give the columns new names
hr_dataset.columns = ['Age', 'Attrition', 'BusinessTravel', 'DailyRate', 'Department', 'DistanceFromHome',
                     'Education', 'EducationField', 'EmployeeNumber', 'EnvironmentSatisfaction', 'Gender',
                     'HourlyRate', 'JobInvolvement', 'JobLevel', 'JobRole', 'JobSatisfaction', 'MaritalStatus',
                     'MonthlyIncome', 'MonthlyRate', 'NumCompaniesWorked', 'OverTime', 'PercentSalaryHike',
                     'PerformanceRating', 'RelationshipSatisfaction', 'StockOptionLevel', 'TotalWorkingYears',
                     'TrainingTimesLastYear', 'WorkLifeBalance', 'YearsAtCompany', 'YearsInCurrentRole',
                     'YearsSinceLastPromotion', 'YearsWithCurrManager']
```

```python
# Eliminate redundant entries
hr_dataset.drop_duplicates(inplace=True)
```

```python
# Sanitize specific columns (if necessary)
# For example, converting 'Gender' column to lowercase
hr_dataset['Gender'] = hr_dataset['Gender'].str.lower()
```

```python
# Eliminate NaN values
hr_dataset.dropna(inplace=True)
```

```python
# Data analysis and visualization
```

```python
# Visualizing Attrition distribution
plt.figure(figsize=(8, 6))
sns.countplot(x='Attrition', data=hr_dataset)
plt.title('Attrition Distribution')
plt.xlabel('Attrition')
plt.ylabel('Count')
plt.show()
```

```python
# Visualizing Age distribution
plt.figure(figsize=(10, 6))
sns.histplot(x='Age', data=hr_dataset, bins=30, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
# Visualizing Attrition vs. MonthlyIncome
plt.figure(figsize=(10, 6))
sns.boxplot(x='Attrition', y='MonthlyIncome', data=hr_dataset)
plt.title('Attrition vs. Monthly Income')
plt.xlabel('Attrition')
plt.ylabel('Monthly Income')
plt.show()
```

```python
# Visualizing Attrition vs. JobSatisfaction
plt.figure(figsize=(10, 6))
sns.boxplot(x='Attrition', y='JobSatisfaction', data=hr_dataset)
plt.title('Attrition vs. Job Satisfaction')
plt.xlabel('Attrition')
plt.ylabel('Job Satisfaction')
plt.show()
```

```python
# Visualizing Attrition vs. WorkLifeBalance
plt.figure(figsize=(10, 6))
sns.boxplot(x='Attrition', y='WorkLifeBalance', data=hr_dataset)
plt.title('Attrition vs. Work Life Balance')
plt.xlabel('Attrition')
plt.ylabel('Work Life Balance')
plt.show()
```

```python
# Visualizing Attrition vs. DistanceFromHome
plt.figure(figsize=(10, 6))
sns.boxplot(x='Attrition', y='DistanceFromHome', data=hr_dataset)
plt.title('Attrition vs. Distance From Home')
plt.xlabel('Attrition')
plt.ylabel('Distance From Home')
plt.show()
```

```python
# Correlation matrix
# Exclude non-numeric columns
numeric_columns = hr_dataset.select_dtypes(include=[np.number]).columns
corr_matrix = hr_dataset[numeric_columns].corr()

plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```
